Q3. What is App routing in Flask? Why do we use app routes?

App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 

Example: In our application, the URL (“/”) is associated with the root URL. So if our site’s domain was www.example.org and we want to add routing to “www.example.org/hello”, we would use “/hello”

Flask’s framework is more explicit than other Python frameworks and easier to learn because it has significantly less base code to implement a simple Web-Application.

App routing is the technique used to map the specific URL with the associated function intended to perform some task. The Latest Web frameworks use the routing technique to help users remember application URLs.

Create dynamic urls using flask url_for function. usernamer(name) function accept a value as argument from the URL. app.route() is used to match url parameters. The url_for()function accepts the function name as first argument . Import flask redirect, url_for from Flask library.

from flask import Flask, redirect, url_for
app = Flask(__name__)

@app.route('/ram')
def ram():
   return 'Welcome Ram'

@app.route('/rahim')
def rahim():
   return 'Welcome Rahim'


@app.route('/username/')
def username(name):
    if name =='ram':
      return redirect(url_for('ram'))
    if name =='rahim':
      return redirect(url_for('rahim'))



if __name__ == '__main__':
   app.run(debug = True)
